```python
import pandas as pd
import re
from difflib import SequenceMatcher
import numpy as np

print("🚀 Starting KRA/KPI mapping process...")

# Create KRA-KPI mappings from File A data
def create_kra_kpi_mappings():
    print("📊 Creating KRA-KPI mappings...")
    mappings = {
        # ANIL GOYAL - HOD Canada Department
        "ANIL GOYAL": [
            {"KRA": "BUSINESS GROWTH & STRATEGY", "KPI": "Monthly enrollment target achievement", "keywords": ["enrollment", "target", "counselling", "new student"]},
            {"KRA": "BUSINESS GROWTH & STRATEGY", "KPI": "Team enrollment target supervision", "keywords": ["team performance", "work allocation", "team supervision"]},
            {"KRA": "BUSINESS GROWTH & STRATEGY", "KPI": "Inquiry-to-enrollment conversion rate", "keywords": ["inquiry", "conversion", "lead", "counselling"]},
            {"KRA": "BUSINESS GROWTH & STRATEGY", "KPI": "Revenue contribution tracking", "keywords": ["revenue", "collection", "business development"]},
            {"KRA": "BUSINESS GROWTH & STRATEGY", "KPI": "Onshore student services revenue growth", "keywords": ["onshore", "toronto office", "student services"]},
            {"KRA": "TEAM LEADERSHIP & MANAGEMENT", "KPI": "Bi-weekly performance reviews", "keywords": ["performance review", "one to one", "team review"]},
            {"KRA": "TEAM LEADERSHIP & MANAGEMENT", "KPI": "Monthly one-on-one meetings", "keywords": ["one-on-one", "feedback", "monthly meeting"]},
            {"KRA": "TEAM LEADERSHIP & MANAGEMENT", "KPI": "Daily team meetings", "keywords": ["daily meeting", "morning meeting", "closure meeting"]},
            {"KRA": "TEAM LEADERSHIP & MANAGEMENT", "KPI": "Team training completion", "keywords": ["training", "internal training", "release training"]},
            {"KRA": "TEAM LEADERSHIP & MANAGEMENT", "KPI": "Team counseling guidance", "keywords": ["coaching", "counseling training", "product training"]},
            {"KRA": "INSTITUTIONAL COORDINATION", "KPI": "University/college liaison activities", "keywords": ["university", "college", "liaison", "engagement"]},
            {"KRA": "INSTITUTIONAL COORDINATION", "KPI": "B2B partner collaboration and visits", "keywords": ["b2b", "partner", "franchise", "agent visit"]},
            {"KRA": "INSTITUTIONAL COORDINATION", "KPI": "Institutional tie-ups and agreements", "keywords": ["tie-up", "agreement", "university tieup"]},
            {"KRA": "INSTITUTIONAL COORDINATION", "KPI": "Application status follow-ups", "keywords": ["application status", "follow-up", "admission followup"]},
            {"KRA": "STRATEGIC COUNSELING", "KPI": "High-potential student counseling sessions", "keywords": ["counseling", "complex case", "refusal", "judicial"]},
            {"KRA": "STRATEGIC COUNSELING", "KPI": "Complex case handling", "keywords": ["complex case", "refusal", "judicial review", "grievance"]},
            {"KRA": "STRATEGIC COUNSELING", "KPI": "Parent meeting coordination", "keywords": ["parent meeting", "onshore parent", "ptm"]},
            {"KRA": "VISA PROCESS OVERSIGHT", "KPI": "Visa success rate", "keywords": ["visa success", "approval rate", "visa analytics"]},
            {"KRA": "VISA PROCESS OVERSIGHT", "KPI": "Visa file review before submission", "keywords": ["file review", "quality assurance", "visa file"]},
            {"KRA": "VISA PROCESS OVERSIGHT", "KPI": "Visa processing timeline compliance", "keywords": ["processing time", "timeline compliance"]},
            {"KRA": "EVENTS & OUTREACH", "KPI": "Monthly seminars/webinars conducted", "keywords": ["seminar", "webinar", "event"]},
            {"KRA": "EVENTS & OUTREACH", "KPI": "Branch visits completion", "keywords": ["branch visit", "institute visit", "franchise visit"]},
            {"KRA": "EVENTS & OUTREACH", "KPI": "Pre-departure seminars and visa fairs", "keywords": ["pre departure", "visa fair", "event participation"]}
        ],

        # RANU SARKAR - Assistant to HOD
        "RANU SARKAR": [
            {"KRA": "EXECUTIVE SUPPORT & COORDINATION", "KPI": "Client meetings arranged with HOD", "keywords": ["client meeting", "meeting arrangement", "appointment"]},
            {"KRA": "EXECUTIVE SUPPORT & COORDINATION", "KPI": "Mobile call handling for HOD", "keywords": ["call handling", "inbound calls", "mobile call"]},
            {"KRA": "EXECUTIVE SUPPORT & COORDINATION", "KPI": "HOD meeting presentations and preparations", "keywords": ["presentation", "meeting preparation", "mom"]},
            {"KRA": "EXECUTIVE SUPPORT & COORDINATION", "KPI": "Document coordination and submission to advisors", "keywords": ["document coordination", "submission", "scanning", "courier"]},
            {"KRA": "COMMUNICATION MANAGEMENT", "KPI": "Student query response time", "keywords": ["query response", "email", "whatsapp", "student query"]},
            {"KRA": "COMMUNICATION MANAGEMENT", "KPI": "Appointment reminders", "keywords": ["appointment reminder", "reminder call", "deadline reminder"]},
            {"KRA": "COMMUNICATION MANAGEMENT", "KPI": "Inter-department coordination", "keywords": ["inter-department", "coordination", "internal team"]},
            {"KRA": "COMMUNICATION MANAGEMENT", "KPI": "WhatsApp group management", "keywords": ["whatsapp group", "broadcast", "group creation"]},
            {"KRA": "OPERATIONAL COORDINATION", "KPI": "Application process coordination", "keywords": ["application process", "form filling", "processing"]},
            {"KRA": "OPERATIONAL COORDINATION", "KPI": "Work allocation within team", "keywords": ["work allocation", "task assignment", "team work"]},
            {"KRA": "OPERATIONAL COORDINATION", "KPI": "Enrollment tracking and reporting", "keywords": ["enrollment tracking", "reporting", "daily report"]}
        ],

        # BHAVIKA SHAH - Coordination Specialist
        "BHAVIKA SHAH": [
            {"KRA": "FOLLOW-UP & COORDINATION", "KPI": "Prospective lead follow-ups", "keywords": ["follow-up", "lead followup", "call", "email", "whatsapp"]},
            {"KRA": "FOLLOW-UP & COORDINATION", "KPI": "Student meeting arrangements", "keywords": ["meeting arrangement", "scheduling", "appointment"]},
            {"KRA": "FOLLOW-UP & COORDINATION", "KPI": "Onshore/offshore student calls", "keywords": ["student call", "calling", "follow up activity"]},
            {"KRA": "FOLLOW-UP & COORDINATION", "KPI": "Student and parent daily meeting coordination", "keywords": ["parent meeting", "student meeting", "ptm"]},
            {"KRA": "APPLICATION & DOCUMENTATION SUPPORT", "KPI": "Form filling assistance", "keywords": ["form filling", "application form", "ds-160"]},
            {"KRA": "APPLICATION & DOCUMENTATION SUPPORT", "KPI": "Document collection coordination", "keywords": ["document collection", "documents collected", "checklist"]},
            {"KRA": "APPLICATION & DOCUMENTATION SUPPORT", "KPI": "Document scanning/uploading", "keywords": ["document scanning", "uploading", "document upload"]},
            {"KRA": "APPLICATION & DOCUMENTATION SUPPORT", "KPI": "Application process updates", "keywords": ["application update", "status update", "real-time update"]},
            {"KRA": "VISA DOCUMENTATION", "KPI": "Student visa checklist explanation", "keywords": ["visa checklist", "document checklist", "visa documentation"]},
            {"KRA": "VISA DOCUMENTATION", "KPI": "Visitor visa checklist for parents", "keywords": ["visitor visa", "super visa", "parent visa"]},
            {"KRA": "VISA DOCUMENTATION", "KPI": "Document verification accuracy", "keywords": ["document verification", "quality check", "file accuracy"]}
        ],

        # MANISH PATIL - Visa Processing Specialist
        "MANISH PATIL": [
            {"KRA": "VISA FILE PROCESSING", "KPI": "Daily visa file completion", "keywords": ["visa file", "student visa", "visitor visa", "super visa"]},
            {"KRA": "VISA FILE PROCESSING", "KPI": "Visa file accuracy", "keywords": ["file accuracy", "error-free", "quality assurance"]},
            {"KRA": "VISA FILE PROCESSING", "KPI": "Processing time efficiency", "keywords": ["processing time", "timeline", "fast-track"]},
            {"KRA": "VISA DOCUMENTATION", "KPI": "Form filling assistance", "keywords": ["form filling", "application form"]},
            {"KRA": "VISA DOCUMENTATION", "KPI": "Student visa checklist delivery", "keywords": ["checklist", "document preparation"]},
            {"KRA": "VISA DOCUMENTATION", "KPI": "Visitor visa checklist for parents", "keywords": ["visitor visa", "parent visa", "super visa"]},
            {"KRA": "VISA DOCUMENTATION", "KPI": "Document verification accuracy", "keywords": ["document verification", "quality check"]},
            {"KRA": "TECHNICAL VISA SUPPORT", "KPI": "Visa file categories expertise", "keywords": ["student visa", "visitor visa", "super visa", "pgwp", "sowp", "pr"]},
            {"KRA": "TECHNICAL VISA SUPPORT", "KPI": "Complex visa case handling", "keywords": ["complex case", "refusal", "judicial review", "appeal"]},
            {"KRA": "TECHNICAL VISA SUPPORT", "KPI": "Quality assurance checks", "keywords": ["quality check", "file review", "compliance"]}
        ],

        # COMMON KRAs for all employees
        "COMMON": [
            {"KRA": "CRM MANAGEMENT", "KPI": "Data accuracy and timely updates in CRM", "keywords": ["crm", "data accuracy", "data management", "crm update"]},
            {"KRA": "CRM MANAGEMENT", "KPI": "Opportunity creation and tracking", "keywords": ["opportunity", "lead tracking", "sales pipeline"]},
            {"KRA": "CRM MANAGEMENT", "KPI": "Follow-up compliance and logging", "keywords": ["follow-up", "activity tracking", "communication records"]},
            {"KRA": "STUDENT COMMUNICATION", "KPI": "Query response", "keywords": ["query response", "email response", "chat response", "phone response"]},
            {"KRA": "STUDENT COMMUNICATION", "KPI": "WhatsApp group management and responses", "keywords": ["whatsapp", "group management", "broadcast"]},
            {"KRA": "STUDENT COMMUNICATION", "KPI": "Email communication effectiveness", "keywords": ["email communication", "professional correspondence", "timely response"]},
            {"KRA": "CAMPUS OPERATIONS", "KPI": "Campus tours conducted", "keywords": ["campus tour", "visitor management", "facility tour"]},
            {"KRA": "CAMPUS OPERATIONS", "KPI": "Visitor management and orientation", "keywords": ["visitor management", "guest management", "orientation"]},
            {"KRA": "CAMPUS OPERATIONS", "KPI": "PTM follow-ups and coordination", "keywords": ["ptm", "parent meeting", "feedback collection"]},
            {"KRA": "CROSS-SELLING ACTIVITIES", "KPI": "Value-added services promotion", "keywords": ["vas", "additional services", "upselling"]},
            {"KRA": "CROSS-SELLING ACTIVITIES", "KPI": "Additional services pitching to students", "keywords": ["service pitching", "product demonstration", "service explanation"]},
            {"KRA": "CROSS-SELLING ACTIVITIES", "KPI": "Revenue generation through ancillary services", "keywords": ["ancillary revenue", "additional income", "service sales"]},
            {"KRA": "COMPLIANCE & TRAINING", "KPI": "Training completion as per schedule", "keywords": ["training", "course completion", "skill development"]},
            {"KRA": "COMPLIANCE & TRAINING", "KPI": "Process adherence and compliance", "keywords": ["compliance", "procedure", "standards"]},
            {"KRA": "COMPLIANCE & TRAINING", "KPI": "Policy implementation", "keywords": ["policy", "guidelines", "rules"]},
            {"KRA": "DEPARTMENT TARGETS", "KPI": "Overall department enrollment targets", "keywords": ["enrollment target", "department target", "team goal"]},
            {"KRA": "DEPARTMENT TARGETS", "KPI": "Visa success rate", "keywords": ["visa success", "approval rate", "department performance"]},
            {"KRA": "DEPARTMENT TARGETS", "KPI": "Student satisfaction ratings", "keywords": ["student satisfaction", "feedback", "service quality"]},
            {"KRA": "DEPARTMENT TARGETS", "KPI": "Team training completion", "keywords": ["team training", "skill development", "group learning"]}
        ]
    }
    return mappings

def calculate_similarity(text1, text2):
    """Calculate similarity between two texts"""
    if not text1 or not text2:
        return 0
    return SequenceMatcher(None, text1.lower(), text2.lower()).ratio()

def find_best_kra_kpi_match(task_title, task_description, assigned_to, mappings):
    """Find the best KRA/KPI match for a task"""
    task_text = f"{task_title} {task_description}".lower()

    best_match = None
    best_score = 0

    # Determine employee type
    employee_key = "COMMON"
    if "manish" in assigned_to.lower():
        employee_key = "MANISH PATIL"
    elif "ranu" in assigned_to.lower():
        employee_key = "RANU SARKAR"
    elif "bhavika" in assigned_to.lower() or "shah" in assigned_to.lower():
        employee_key = "BHAVIKA SHAH"
    elif "anil" in assigned_to.lower() or "goyal" in assigned_to.lower():
        employee_key = "ANIL GOYAL"

    # Check employee-specific mappings first
    if employee_key in mappings:
        for mapping in mappings[employee_key]:
            # Check keyword matches
            keyword_score = 0
            for keyword in mapping.get('keywords', []):
                if keyword in task_text:
                    keyword_score += 1

            # Calculate overall score
            kpi_text = mapping['KPI'].lower()
            kra_text = mapping['KRA'].lower()

            kpi_similarity = calculate_similarity(task_text, kpi_text)
            kra_similarity = calculate_similarity(task_text, kra_text)
            keyword_similarity = min(keyword_score * 0.2, 1.0)  # Cap keyword influence

            total_score = (kpi_similarity * 0.5 + kra_similarity * 0.3 + keyword_similarity * 0.2)

            if total_score > best_score:
                best_score = total_score
                best_match = mapping

    # If no good employee-specific match, check common mappings
    if best_score < 0.3 and "COMMON" in mappings:
        for mapping in mappings["COMMON"]:
            keyword_score = 0
            for keyword in mapping.get('keywords', []):
                if keyword in task_text:
                    keyword_score += 1

            kpi_text = mapping['KPI'].lower()
            kra_text = mapping['KRA'].lower()

            kpi_similarity = calculate_similarity(task_text, kpi_text)
            kra_similarity = calculate_similarity(task_text, kra_text)
            keyword_similarity = min(keyword_score * 0.2, 1.0)

            total_score = (kpi_similarity * 0.5 + kra_similarity * 0.3 + keyword_similarity * 0.2)

            if total_score > best_score:
                best_score = total_score
                best_match = mapping

    return best_match

# Load File B data
print("📖 Reading Excel file...")
file_b_df = pd.read_excel('Canada B2C - Head Office All Task_OCT27-NOV27_export.xlsx', sheet_name='ho-admin-2025-11-27T09_27_24.17')

# Create mappings
mappings = create_kra_kpi_mappings()

# Process all tasks and assign KRA/KPI
kra_list = []
kpi_list = []

print("🔄 Processing tasks and assigning KRA/KPI mappings...")
print(f"Total tasks to process: {len(file_b_df)}")

for index, row in file_b_df.iterrows():
    assigned_to = str(row['Assigned To']) if pd.notna(row['Assigned To']) else ""
    title = str(row['Title']) if pd.notna(row['Title']) else ""
    description = str(row['Description']) if pd.notna(row['Description']) else ""

    match = find_best_kra_kpi_match(title, description, assigned_to, mappings)

    if match:
        kra_list.append(match['KRA'])
        kpi_list.append(match['KPI'])
    else:
        kra_list.append("")
        kpi_list.append("")

    # Progress indicator
    if (index + 1) % 50 == 0:
        print(f"   Processed {index + 1} tasks...")

# Add KRA and KPI columns to the dataframe
file_b_df['KRA'] = kra_list
file_b_df['KPI'] = kpi_list

print("✅ Processing complete!")

# Save the updated file
output_filename = "Canada_B2C_Head_Office_All_Task_with_KRA_KPI.xlsx"
file_b_df.to_excel(output_filename, index=False)

print(f"💾 Updated file saved as: {output_filename}")

# Display some statistics
assigned_kra_count = len([kra for kra in kra_list if kra])
assigned_kpi_count = len([kpi for kpi in kpi_list if kpi])

print(f"\n📊 Mapping Statistics:")
print(f"   Total tasks processed: {len(file_b_df)}")
print(f"   Tasks with KRA assigned: {assigned_kra_count} ({assigned_kra_count/len(file_b_df)*100:.1f}%)")
print(f"   Tasks with KPI assigned: {assigned_kpi_count} ({assigned_kpi_count/len(file_b_df)*100:.1f}%)")

# Show sample of mapped tasks
print(f"\n🔍 Sample Mapped Tasks:")
sample_data = file_b_df[['Assigned To', 'Title', 'KRA', 'KPI']].head(10)
for i, row in sample_data.iterrows():
    print(f"   👤 {row['Assigned To']}: {row['Title'][:50]}...")
    print(f"      KRA: {row['KRA']}")
    print(f"      KPI: {row['KPI']}")
    print()
```
